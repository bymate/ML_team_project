```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import time

def script_scroll_v2(url):
    driver = webdriver.Chrome()
    driver.get(url)

    expand_button = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.ID, 'expand')))
    expand_button.click()

    button = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, '//button[contains(., "스크립트 표시")]')))
    button.click()

    WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.ID, 'secondary')))
    time.sleep(3)

    # 페이지 소스 가져오기
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')

    # 자막 세그먼트 모두 가져오기
    segments = soup.find_all("div", class_="segment style-scope ytd-transcript-segment-renderer")

    # 자막 시간과 텍스트 추출
    timestamps = []
    texts = []

    for segment in segments:
        timestamp = segment.find("div", class_="segment-timestamp style-scope ytd-transcript-segment-renderer").get_text(strip=True)  
        text = segment.find("yt-formatted-string", class_="segment-text style-scope ytd-transcript-segment-renderer").get_text(strip=True) 
        timestamps.append(timestamp)
        texts.append(text)

    # 결과 출력 (타임스탬프와 자막 텍스트)
    # for time, text in zip(timestamps, texts):
    #     print(f"{time} - {text}")

    text_str = '\n'.join(texts) # 스크립트 하나의 str로
    driver.quit()
    return texts
```

```python
url = r"https://www.youtube.com/watch?v=BruwoWzx3vA"
text_content = script_scroll_v2(url)
print(text_content)
print(type(text_content))
```

```python
# 한글 버전 - 자동 완성 자막

from youtube_transcript_api import YouTubeTranscriptApi
from urllib.parse import urlparse, parse_qs

def parse_script(url):
    parsed_url = urlparse(url)
    video_id = parse_qs(parsed_url.query).get('v')
 
    srt = YouTubeTranscriptApi.get_transcript(video_id[0], languages=['ko'])
    text_list = [item['text'] for item in srt]

    text_str = '\n'.join(text_list) # 스크립트 하나의 str로
    return text_list

url = r"https://www.youtube.com/watch?v=BruwoWzx3vA"
text_list = parse_script(url)

print(text_list)
print(type(text_list))
```
